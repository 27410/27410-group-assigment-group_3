# Selection and assessment of existing GSM
Our choice of organism, Saccharomyces cerevisiae, is a very common and well-researched organism, thus multiple GSMs exist. From BIGG and EMBL-EBI's BioModels we found four candidate GSMs: iFF708, iMM904, iND750 and yeast-GEM (version 8.6.2 newest on GitHub).

Using `memote`, we can assess different quality measures of the models, such as stochiometry, annotation and reaction/metabolite statistics.

```python
# Testing iFF708
%%time
!memote report snapshot models/iFF708.xml --filename models/memote/iFF708.html
```

```python
# Testing iMM904
%%time
!memote report snapshot models/iMM904.xml --filename models/memote/iMM904.html
```

```python
# Testing iND750
%%time
!memote report snapshot models/iND750.xml --filename models/memote/iND750.html
```

```python
# Testing yeast8.6.2
%%time
!memote report snapshot models/yeast-GEM.xml --filename models/memote/yeast-GEM.html
```

The results of the `memote` runs can be seen in the `models/memote` folder.

The results are summarised in the table below

| Measure | iMM904 | yeast8.6.2 | iFF708 | iND750 |
| ---- | ---- | - | - | - |
| Total Metabolites | 1,226 | 2,744 | 796 | 1,059 |
| Total Reactions | 1,577 | 4,063 | 1,379 | 1.266 |
| Total Genes | 905 | 1,160 | 619 | 750 |
|  Stochiometric Consistency  | 100.0% | 0.0% | 0.0% | 100.0% |
|  Mass Balance  | 96.0% | 93.7% | 0.0% | 97.3% |
|  Charge Balance  | 98.5% | 98.2% | 100.0% | 100.0% |
|  Metabolite Connectivity  | 100.0% | 100.0% | 100.0% | 100.0% |
|  Unbounded Flux In Default Medium | 76.2% | 58.8% | 71.7% | 83.0% |
|  Metabolite Annotation | 80% | 68% | 25% | 80% |
|  Reaction Annotation | 82% | 65% | 25% | 83% |
|  Gene Annotation | 43% | 54% | 0% | 43% |
|  **Total score** | 85% | 68% | 19% | 86% |

We decided that the stochiometric consistency was a more important parameter than the number of reactions, which left us with either iMM904 or iND750. Since the scores are so close, we decided to go for iMM904, as it contains the most reactions, metabolites and genes.

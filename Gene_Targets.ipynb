Evaluation of possible gene targets for overexpression and knockouts. The evaluation is done using Flux Scanning based Enforced Objective Flux (FSEOF) from the CAMEO package.  


Flux through heterologous pathway is dependent on disrupting flux through the natural pathway to Ergosterol
erg6 + erg5 deletion, either setting flux = 0 or by cobra.flux_analysis_single_gene_deletion

```python
from cobra.io import read_sbml_model

yeast_model = read_sbml_model('models/iMM904.xml')
yeast_model
```

```python
print('complete model: ', yeast_model.optimize())
with yeast_model:
    yeast_model.genes.YMR015C.knock_out()
    yeast_model.genes.YML008C.knock_out()
    print('ERG5 and ERG6 knocked out: ', yeast_model.optimize())
```

The biomass of the WT strain and the strain with a knockout in ERG5 and ERG6 is the same meaning that it should be appropiate to implement these deletions for driving the heterologous pathway for production of progesterone.

OptForce from CAMEO package is used to analyse what strain engineering strategy can be done in order to optimize flux to progesterone.

```python
from cameo.strain_design import OptGene
optgene = OptGene(yeast_model)
result = optgene.run(target=yeast_model.reactions.EX_zymst_e, biomass=yeast_model.reactions.BIOMASS_SC5_notrace, substrate=yeast_model.metabolites.zymst_c, max_evaluations=10, plot=False)
```
